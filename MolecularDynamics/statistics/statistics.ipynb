```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
energies = pd.read_csv('energies.csv').set_index("time")
energies.head()
fig, axes= plt.subplots(nrows=3, figsize=(15, 10))
energies[['kinetic_energy_pp','potential_energy_pp','total_energy_pp']].plot(ax = axes[0])
energies['total_energy_pp'].plot(ax = axes[1])
energies['temperature'].plot(ax = axes[2])
```
