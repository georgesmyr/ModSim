```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
perf = pd.read_csv("performance0.csv")
for i in range(8):
    perf = perf.add(pd.read_csv("performance" + str(i + 1) + ".csv"))
perf = perf.divide(9)
```

```python
perfn = perf[perf.CostFunction != 0]
perfn = perfn[perfn.CostFunction == perfn.CostFunction.min()]
perfn[perfn.Steps == perfn.Steps.min()]
```

```python
def plot_energy(name):
    df = pd.read_csv(name + ".csv")
    fig, ax = plt.subplots()
    ax.plot(df.Energy)
    ax.set_xlabel('Steps')
    ax.set_ylabel('Cost Function')
    ax.set_title('Cost Function - Time')
    plt.savefig(name + '.png')
```

```python
plot_energy('energies1')
```

```python
plot_energy('energies_easy')
```

```python
plot_energy('energies_medium')
```
