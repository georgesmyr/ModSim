```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
averagePressures = []
averageMus = []
densities = []
for i in range(60):
    meas = pd.read_csv("measurments" + str(i + 1) + ".csv")
    averagePressures.append(meas.AveragePressure.mean())
    averageMus.append(meas.MuExcess.mean())
    densities.append((i+1) * 0.02)
    
df_dict = {"Average Pressure" : averagePressures, "Average Mus" : averageMus}
df = pd.DataFrame(df_dict, index = densities)
```

```python
df["Average Pressure"].plot()
```

```python
df["Average Mus"].plot()
```
